# Cohere Results

Zero-shot results using Cohere's chat API to evaluate misinformation, using the X-Fact dataset.

```python
import pandas as pd
df = pd.read_csv("x-fact/data/x-fact-including-en/train.all.tsv", delimiter="\t", on_bad_lines="skip")

# this following line might work if the above line has a bug
# df = pd.read_csv("../x-fact/data/x-fact-including-en/train.all.tsv", delimiter="\t", on_bad_lines="skip")
```

```python
COHERE_API_KEY="Get your Cohere API key at this link: https://dashboard.cohere.com/api-keys"
```

```python
import cohere
co = cohere.Client(COHERE_API_KEY)
```

```python
import time
from tqdm import trange

cohere_resp = []

for i in trange(20):
    tmp = {"cohere_resp":None, "index":i, "label":df["label"][i]}
    claim = df["claim"][i]
    response = co.chat(
        message=f"Is the following claim true: {claim}",
        connectors=[{"id": "web-search"}]
    )
    tmp["cohere_resp"] = response.text
    cohere_resp.append(tmp)
    time.sleep(1)
```

```python
import termcolor
df_results = pd.DataFrame(cohere_resp)
for i in range(df_results.shape[0]):
    print("Index: "+str(i), end=".\n")
    print(termcolor.colored("True label:\t","magenta")+ termcolor.colored(df_results["label"][i],"blue"))
    print(termcolor.colored("LLM response:\t", "cyan")+df_results["cohere_resp"][i])
    print("\n---\n")
```
